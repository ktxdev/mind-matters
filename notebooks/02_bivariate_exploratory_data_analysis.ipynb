# Bivariate Analysis

## Gender vs. Depression

```python
visualizer.plot('Gender', 'Depression')
```

### Insights:
- Both males and females experience depression, but the absolute count of males experiencing depression is higher, likely reflecting the sample size rather than a gender-based predisposition
- The gender disparity could also reflect societal expectations and stigma, influencing self-reporting tendencies.

## Age vs. Depression

```python
visualizer.plot('Age', 'Depression')
```

### Insights:
- Younger individuals are more prone to depression, as indicated by the lower median and narrower interquartile range for individuals with depression compared to those without it
- Older individuals tend to report less depression, possibly due to increased stability or resilience developed over time.

## City vs. Depression

```python
visualizer.plot('City', 'Depression')
```

### Insights:
- The distribution of depression varies across cities. Certain cities, like Indore, appear to have a larger proportion of individuals without depression. In contrast, other cities, such as Ludhiana or Kolkata, show a relatively higher proportion of individuals experiencing depression
- Urban factors such as socio-economic conditions, access to mental health resources, and work environments might influence these variations.

## Working Professional or Student vs Depression

```python
visualizer.plot('Working Professional or Student', 'Depression')
```

### Insights:
- The majority of the dataset consists of working professionals compared to students
- Depression (indicated in orange) appears to be more prevalent among students proportionally, even though working professionals are numerically dominant in the dataset. This could point to the specific challenges faced by students, such as academic pressure and career uncertainties

## Profession vs. Depression

```python
visualizer.plot('Profession', 'Depression')
```

### Insights:
- The student group shows a significant proportion of depression. This indicates that students are particularly vulnerable to depression compared to other professions
- Teachers have a high count overall but show a smaller proportion of depression compared to their "Not Depressed" group. This suggests that  while teachers experience some level of stress, it may not be as significant a factor for depression relative to their population size

## Academic Pressure vs. Depression

```python
visualizer.plot('Academic Pressure', 'Depression')
```

### Insights:
- Both depressed and non-depressed individuals exist in this category, but the proportion of depressed individuals appears relatively small
- For students (categories 15), there seems to be a positive relationship between academic pressure and depression:
    - Higher levels of academic pressure (e.g., 4.0, 5.0) are associated with a higher proportion of depressed individuals
    - Lower academic pressure levels (e.g., 1.0, 2.0) have a smaller proportion of depressed individuals
- This suggests that academic pressure is a significant factor influencing depression among students
